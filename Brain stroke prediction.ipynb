### 3.0-15(child),15-30(teenage),30-50(Adult),50-above(old age)create a new column by category.
 

```python
data
```

```python
AGE1=[]
for x in data["age"]:
    if x>0 and x<=15:
        AGE1.append("Child")
    elif x>15 and x<=30:
        AGE1.append("Teenage")
    elif x>30 and x<=50:
        AGE1.append("Adult")
    elif x>50:
        AGE1.append("Old age")
print(AGE1)
```

```python
data.insert(2,'Age',AGE1)
```
